```python
from sklearn.preprocessing import PowerTransformer
```

```python
x
```

```python
x.shape
```

```python
y
```

```python
y.shape
```

```python
pt=PowerTransformer(method="yeo-johnson")
X_power=pt.fit_transform(x)
```

```python
df1=pd.DataFrame(X_power,columns=x.columns)
```

```python
df1.skew()
```

last_rech_date_da has high skewness we have to drop it

```python
df1.drop(['last_rech_date_da'],axis=1,inplace=True)
```

```python
df1.skew()
```

still some columns show skewness,remove it by sqrt

```python
from scipy import sqrt
```

```python
df1["daily_decr30"]=np.sqrt(df1["daily_decr30"])
df1["daily_decr90"]=np.sqrt(df1["daily_decr90"])
df1["medianmarechprebal30"]=np.sqrt(df1["medianmarechprebal30"])
df1["medianmarechprebal90"]=np.sqrt(df1["medianmarechprebal90"])
df1["cnt_da_rech90"]=np.sqrt(df1["cnt_da_rech90"])
df1["medianamnt_loans30"]=np.sqrt(df1["medianamnt_loans30"])
df1["medianamnt_loans90"]=np.sqrt(df1["medianamnt_loans90"])
```

# MULTICOLLINEARITY

USING VIF FOR CHECKING MULTICOLLINEARITY

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def vif_calc():
    vif=pd.DataFrame()
    vif['VIF Factor']=[variance_inflation_factor(x1.values,i)for i in range(x1.shape[1])]
    vif['Features']=x1.columns
    print(vif)
```
